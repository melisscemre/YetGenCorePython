# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
class Kumanda:
    def __init__(self):
        self.tv_acik= False
        self.ses= 10 
        self.kanal_listesi= ["SZC", "Onlar", "Netflix"]
        self.anlik_kanal= self.kanal_listesi[0]
    
    def tv_ac(self):
        if self.tv_acik:
            print("TV zaten açık.")
        else:
            self.tv_acik = True
            print("TV açıldı.")

    def tv_kapat(self):
        if not self.tv_acik:
            print("TV zaten kapalı.")
        else:
            self.tv_acik= False
            print("TV kapandı.")
    
    def ses_ayarlari(self):
        while True:
            secim = input("Sesi arttırmak için '+', azaltmak için '-', çıkmak için 'q': ")
            if secim == "+":
                if self.ses < 100:
                    self.ses += 1
                    print("Ses seviyesi:" , self.ses)
                else: print("Ses maksimumda.")
            elif secim == "-":
                if self.ses > 0:
                    self.ses -= 1
                    print("Ses seviyesi:" , self.ses)
                else: print("Ses minimumda.")
            elif secim == "q":
                print("Ses ayarından çıkılıyor.")
            else:
                print("Geçersiz giriş.")

    def kanal_ekle(self):
        yeni_kanal= input("eklemek istediğiniz kanalın adını yazınız: ")
        self.kanal_listesi.append(yeni_kanal)
        print(f" {yeni_kanal} başarıyla eklendi.")

    def kanal_degistir(self):
        print("Kanallar:", ", ".join(self.kanal_listesi))
        secim= input("geçmek istediğiniz kanalın adını yazınız: ")
        if secim in self.kanal_listesi: 
            self.anlik_kanal = secim
            print("Kanal değiştirildi:" , self.anlik_kanal)
        else: 
            print("Kanal bulunamdı.")

    def kanal_sayisi(self):
        print(f"Toplam kanal sayısı: {len(self.kanal_listesi)}")
    
    def tv_bilgileri(self):
        print("\n Televizyon Bilgileri")
        print("Durum:", "Açık" if self.tv_acik else "Kapalı")
        print("Ses Seviyesi:", ", ".join(self.ses_ayarlari))
        print("Aktif Kanal:", self.anlik_kanal)
        print("*" * 30)

kumanda = Kumanda()

def menu():
      print("""
===== KUMANDA MENÜSÜ =====
1. TV Aç
2. TV Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Göster
6. Kanal Değiştir
7. TV Bilgilerini Göster
q. Çıkış
""")
      

while True:
    menu()
    secim = input("Lütfen bir işlem seçiniz: ")
    if secim == "1":
        kumanda.tv_ac()
    elif secim == "2":
        kumanda.tv_kapat()
    elif secim == "3":
        kumanda.ses_ayarlari()
    elif secim == "4":
        kumanda.kanal_ekle()
    elif secim == "5":
        kumanda.kanal_sayisi()
    elif secim == "6":
        kumanda.kanal_degistir()
    elif secim == "7":
        kumanda.tv_bilgileri()
    elif secim == "q":
        print("📴 Kumanda kapatılıyor... Güle güle! 👋")
        break
    else:
        print("⚠️ Geçersiz seçim. Lütfen tekrar deneyin.")
    
```
