# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self):
        self.tv_acik_mi = False
        self.ses = 10
        self.kanallar = ["TRT"]
        self.mevcut_kanal = self.kanallar[0]

    def tv_ac(self):
        if not self.tv_acik_mi:
            self.tv_acik_mi = True
            print("Televizyon açılıyor...")
            time.sleep(1)
            print("Televizyon açık.")
        else:
            print("Televizyon zaten açık.")

    def tv_kapat(self):
        if self.tv_acik_mi:
            print("Televizyon kapanıyor...")
            time.sleep(1)
            self.tv_acik_mi = False
            print("Televizyon kapalı.")
        else:
            print("Televizyon zaten kapalı.")

    def ses_ayar(self):
        if not self.tv_acik_mi:
            print("Lütfen önce televizyonu açın.")
            return

        while True:
            cevap = input("Sesi artırmak için '+', azaltmak için '-', çıkmak için 'q': ")
            if cevap == '+':
                if self.ses < 100:
                    self.ses += 1
                    print(f"Ses: {self.ses}")
                else:
                    print("Ses zaten maksimumda.")
            elif cevap == '-':
                if self.ses > 0:
                    self.ses -= 1
                    print(f"Ses: {self.ses}")
                else:
                    print("Ses zaten minimumda.")
            elif cevap == 'q':
                break
            else:
                print("Geçersiz tuşlama.")

    def kanal_ekle(self, kanal_ismi):
        if kanal_ismi not in self.kanallar:
            self.kanallar.append(kanal_ismi)
            print(f"{kanal_ismi} kanalı eklendi.")
        else:
            print(f"{kanal_ismi} zaten mevcut.")

    def kanal_sayisi(self):
        print(f"Toplam {len(self.kanallar)} kanal var.")

    def rastgele_kanal(self):
        if not self.tv_acik_mi:
            print("Lütfen önce televizyonu açın.")
            return

        self.mevcut_kanal = random.choice(self.kanallar)
        print(f"Şu anki kanal: {self.mevcut_kanal}")

    def tv_bilgileri(self):
        print("\n--- TV Bilgileri ---")
        print(f"Televizyon durumu: {'Açık' if self.tv_acik_mi else 'Kapalı'}")
        print(f"Ses seviyesi: {self.ses}")
        print(f"Kanallar: {self.kanallar}")
        print(f"Mevcut kanal: {self.mevcut_kanal}")
        print("--------------------\n")


# Kullanıcı etkileşimli menü
kumanda = Kumanda()

while True:
    print("""
    --- Kumanda Menüsü ---
    1 - TV Aç
    2 - TV Kapat
    3 - Ses Ayarları
    4 - Kanal Ekle
    5 - Kanal Sayısını Göster
    6 - Rastgele Kanala Geç
    7 - Televizyon Bilgilerini Göster
    q - Çıkış
    """)
    secim = input("Bir işlem seçin: ")

    if secim == '1':
        kumanda.tv_ac()
    elif secim == '2':
        kumanda.tv_kapat()
    elif secim == '3':
        kumanda.ses_ayar()
    elif secim == '4':
        kanal_ismi = input("Eklenecek kanal ismini girin: ")
        kumanda.kanal_ekle(kanal_ismi)
    elif secim == '5':
        kumanda.kanal_sayisi()
    elif secim == '6':
        kumanda.rastgele_kanal()
    elif secim == '7':
        kumanda.tv_bilgileri()
    elif secim == 'q':
        print("Kumanda kapatılıyor...")
        break
    else:
        print("Geçersiz işlem, tekrar deneyin.")
```
