# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
def vki_hesapla(kilo, boy):
    return kilo / (boy ** 2)

def vki_siniflandir(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 25:
        return "Normal kilolu"
    elif 25 <= vki < 30:
        return "Fazla kilolu"
    elif 30 <= vki < 40:
        return "Obez"
    else:
        return "İleri derecede obez (morbid obez)"

def veri_girisi():
    kayitlar = []
    try:
        adet = int(input("Kaç kişinin verisi girilecek? "))
    except ValueError:
        print("❌ Lütfen geçerli bir sayı giriniz.")
        return

    for i in range(adet):
        print(f"\n📌 {i+1}. kişi için veri giriniz:")
        ad_soyad = input("Ad Soyad: ")
        try:
            yas = int(input("Yaş: "))
            boy = float(input("Boy (metre): "))
            kilo = float(input("Kilo (kg): "))
        except ValueError:
            print("❌ Hatalı giriş. Bu kişi atlanıyor.")
            continue

        vki = vki_hesapla(kilo, boy)
        sinif = vki_siniflandir(vki)

        kayit = f"{ad_soyad}, Yaş: {yas}, Boy: {boy} m, Kilo: {kilo} kg, VKİ: {vki:.2f}, Sınıf: {sinif}"
        kayitlar.append(kayit)

        print(f"✅ VKİ: {vki:.2f} - {sinif}")

    # Dosyaya kayıt
    with open("vki_kayitlari.txt", "a", encoding="utf-8") as dosya:
        dosya.write("\n".join(kayitlar) + "\n")

    print("\n📁 Tüm veriler 'vki_kayitlari.txt' dosyasına kaydedildi.")

# Programı başlat
veri_girisi()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import datetime

# Smoothie menüsü ve fiyatları
menu = {
    "Çilek": 15,
    "Muz": 12,
    "Ispanak": 10,
    "Yaban Mersini": 20,
    "Protein Tozu": 25,
    "Badem Sütü": 8
}

# Satın alma geçmişi
log_file = "yetbar_log.txt"

def menu_goster():
    print("\n📋 YETBAR MENÜSÜ:")
    for malzeme, fiyat in menu.items():
        print(f"- {malzeme}: {fiyat} TL")
    print()

def satin_al():
    toplam_fiyat = 0
    secilenler = []

    print("\nLütfen smoothie'nize eklemek istediğiniz malzemeleri girin.")
    print("İşlemi bitirmek için 'q' ya basın.\n")

    while True:
        secim = input("Malzeme: ").title()

        if secim == 'Q':
            break
        elif secim in menu:
            secilenler.append(secim)
            toplam_fiyat += menu[secim]
            print(f"✔ {secim} eklendi. Şu anki toplam: {toplam_fiyat} TL\n")
        else:
            print("❌ Menüde böyle bir ürün yok. Lütfen tekrar deneyin.\n")

    if secilenler:
        zaman = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        log_satiri = f"{zaman} | Seçilenler: {', '.join(secilenler)} | Toplam: {toplam_fiyat} TL\n"
        with open(log_file, "a") as dosya:
            dosya.write(log_satiri)
        print(f"\n🧾 Siparişiniz alındı! Toplam tutar: {toplam_fiyat} TL")
    else:
        print("⚠️ Hiçbir malzeme seçilmedi. İşlem iptal edildi.")

def log_goster():
    print("\n📁 Satın Alma Geçmişi:\n")
    try:
        with open(log_file, "r") as dosya:
            print(dosya.read())
    except FileNotFoundError:
        print("🔍 Henüz kayıtlı bir satın alma yok.")

def ana_menu():
    while True:
        print("\n🎉 YETBAR - Smoothie Bank'e Hoş Geldiniz 🎉")
        print("1 - Menü Göster")
        print("2 - Smoothie Satın Al")
        print("3 - Satın Alma Geçmişini Göster")
        print("Q - Çıkış")

        secim = input("Seçiminiz: ").strip().upper()

        if secim == '1':
            menu_goster()
        elif secim == '2':
            satin_al()
        elif secim == '3':
            log_goster()
        elif secim == 'Q':
            print("👋 Görüşmek üzere!")
            break
        else:
            print("❗ Geçersiz seçim. Lütfen tekrar deneyin.")

# Programı çalıştır
ana_menu()
```
