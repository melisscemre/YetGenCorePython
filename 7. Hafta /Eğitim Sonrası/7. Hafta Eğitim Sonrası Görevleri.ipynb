 **Bu bir veri giriÅŸ platformu olacak**


 * kullanÄ±cÄ±ya kaÃ§ giriÅŸ yapacaÄŸÄ±nÄ± soracaÄŸÄ±z
* kullanÄ±cÄ±dan isim-soyisim, yaÅŸ, boy, kilo bilgilerini alarak VKÄ° hesaplayacaÄŸÄ±z
* ardÄ±ndan bunlarÄ± sÄ±nÄ±flara gÃ¶re kaydedeceÄŸiz bir dosyaya kaydedeceÄŸiz


SÄ±nÄ±flar ise ÅŸu ÅŸekilde: 
* 18, 5 kg/m.'nin altÄ±nda olanlar: ZayÄ±f
* 18.5 â€“ 24, 9 kg/m. arasÄ±nda olanlar: Normal kilolu
* 25 â€“ 29, 9 kg/m. arasÄ±nda olanlar: Fazla kilolu
* 30 â€“ 39, 9 kg/m. arasÄ±nda olanlar: Obez
* 40 kg/m.'nin Ã¼zerinde olanlar: Ä°leri derecede obez (morbid obez), olarak gÃ¶rÃ¼lÃ¼r.

```python
# 1.Problemin Ã‡Ã¶zÃ¼mÃ¼nÃ¼ Buraya YazÄ±nÄ±z
def vki_hesapla(kilo, boy):
    return kilo / (boy ** 2)

def vki_siniflandir(vki):
    if vki < 18.5:
        return "ZayÄ±f"
    elif 18.5 <= vki < 25:
        return "Normal kilolu"
    elif 25 <= vki < 30:
        return "Fazla kilolu"
    elif 30 <= vki < 40:
        return "Obez"
    else:
        return "Ä°leri derecede obez (morbid obez)"

def veri_girisi():
    kayitlar = []
    try:
        adet = int(input("KaÃ§ kiÅŸinin verisi girilecek? "))
    except ValueError:
        print("âŒ LÃ¼tfen geÃ§erli bir sayÄ± giriniz.")
        return

    for i in range(adet):
        print(f"\nğŸ“Œ {i+1}. kiÅŸi iÃ§in veri giriniz:")
        ad_soyad = input("Ad Soyad: ")
        try:
            yas = int(input("YaÅŸ: "))
            boy = float(input("Boy (metre): "))
            kilo = float(input("Kilo (kg): "))
        except ValueError:
            print("âŒ HatalÄ± giriÅŸ. Bu kiÅŸi atlanÄ±yor.")
            continue

        vki = vki_hesapla(kilo, boy)
        sinif = vki_siniflandir(vki)

        kayit = f"{ad_soyad}, YaÅŸ: {yas}, Boy: {boy} m, Kilo: {kilo} kg, VKÄ°: {vki:.2f}, SÄ±nÄ±f: {sinif}"
        kayitlar.append(kayit)

        print(f"âœ… VKÄ°: {vki:.2f} - {sinif}")

    # Dosyaya kayÄ±t
    with open("vki_kayitlari.txt", "a", encoding="utf-8") as dosya:
        dosya.write("\n".join(kayitlar) + "\n")

    print("\nğŸ“ TÃ¼m veriler 'vki_kayitlari.txt' dosyasÄ±na kaydedildi.")

# ProgramÄ± baÅŸlat
veri_girisi()
```
